```python
# rename ground truth masks name in truck spinnerf directory for easier evaluation.

import os
import re

def rename_files_truck_spinnnerf(directory, rename_dir):
    # Regular expression pattern to match files in the format 'a_b.png'
    pattern = re.compile(r'^[^_]+_(.+)\.png$')

    # Iterate through all files in the specified directory
    for filename in os.listdir(directory):
        # Check if the filename matches the pattern
        match = pattern.match(filename)
        if match:
            
            new_filename = f"{match.group(1)}.png"
           
            old_file = os.path.join(directory, filename)
            new_file = os.path.join(rename_dir, new_filename)
            # Rename the file
            os.rename(old_file, new_file)
            print(f"Renamed '{filename}' to '{new_filename}'")


directory_path = '/home/jainuma1/PhD/rp1/gaussian_cut/Spinnerf/Truck (Tanks & Temples)'
rename_dir = '/home/jainuma1/PhD/rp1/gaussian_cut/Spinnerf/Truck_renamed'
os.makedirs(rename_dir, exist_ok=True)
rename_files_truck_spinnnerf(directory_path, rename_dir)


```
